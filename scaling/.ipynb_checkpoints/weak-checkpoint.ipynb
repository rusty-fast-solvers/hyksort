```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
weak = pd.read_csv('hyksort_scaling.csv')

weak.keys()
```

```python
# Calculate performance metrics
fig, ax = plt.subplots()

n_points = 1e8

ax.plot(weak['n_processes']*n_points, weak['runtime(s)'],  '-o', label='Total Runtime')

ax.set_ylabel('Runtume (s)')
ax.set_xlabel('Number of Points')

plt.legend()
plt.show()
```
