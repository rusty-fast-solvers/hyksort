```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
weak = pd.read_csv('hyksort_scaling2.csv')

weak.keys()
```

```python
# Calculate performance metrics
fig, ax = plt.subplots()

n_points = 1e8

ax.set_xscale('log')
ax.plot(weak['n_processes']*n_points, weak['runtime(s)'],  '-o', label='hyksort')

ax.set_title('HykSort Scaling')
ax.set_ylabel('Runtume (s)')
ax.set_xlabel('Number of Points')

plt.legend()
plt.show()
```

```python
# Calculate performance metrics
fig, ax = plt.subplots()


# ax.set_yscale('log')
# ax.set_xscale('log')
ax.plot(weak['n_processes']*n_points, weak['peak_memory_per_node(K)']/1e6,  '-o', label='memory')

ax.set_ylabel('Peak Memory Per Node (GB)')
ax.set_xlabel('Number of Points')

# plt.legend()
plt.show()
```
