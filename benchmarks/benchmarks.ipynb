```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Compare Hyksort (Rust) to USort (C++) performance

```python
df = pd.read_csv('timing.csv')
rust = df[df['language'] == 'rust']
cpp = df[df['language'] == 'cpp']
```

```python
cols = ['k', 'nprocs', 'nparticles', 'wall', 'local_sort', 'communication']
rust = rust[cols]
cpp = cpp[cols]
```

```python
fig, ax = plt.subplots()

k2_rust = rust[rust['k'] == 2]
k2_cpp = cpp[cpp['k'] == 2]

ax.plot(k2_rust['nprocs'], k2_rust['wall'], '-o', label='Rust', )
ax.plot(k2_cpp['nprocs'], k2_cpp['wall'], '-o', label='CPP')
ax.set_title('Wall Time Weak Scaling (OMP_THREADS=1, K=2)')
ax.set_ylabel('Wall Time (ms)')
ax.set_xlabel('Number of MPI Processes')
ax.legend()
```

```python
fig, ax = plt.subplots()

k2_rust = rust[rust['k'] == 2]
k2_cpp = cpp[cpp['k'] == 2]

ax.plot(k2_rust['nprocs'], k2_rust['local_sort'], '-o', label='Rust', )
ax.plot(k2_cpp['nprocs'], k2_cpp['local_sort'], '-o', label='CPP')
ax.set_title('Local Sort Weak Scaling (OMP_THREADS=1, K=2)')
ax.set_ylabel('Wall Time (ms)')
ax.set_xlabel('Number of MPI Processes')
ax.legend()
```

```python
fig, ax = plt.subplots()

k2_rust = rust[rust['k'] == 2]
k2_cpp = cpp[cpp['k'] == 2]

ax.plot(k2_rust['nprocs'], k2_rust['communication'], '-o', label='Rust', )
ax.plot(k2_cpp['nprocs'], k2_cpp['communication'], '-o', label='CPP')
ax.set_title('Comunication Weak Scaling (OMP_THREADS=1, K=2)')
ax.set_ylabel('Wall Time (ms)')
ax.set_xlabel('Number of MPI Processes')
ax.legend()
```
